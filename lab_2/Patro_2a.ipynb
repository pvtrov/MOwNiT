# Interpolations

```python
def lagrange(x, y):
    poly = np.polynomial.Polynomial([0])
    for i in range(x.shape[0]):
        poly += y[i] * np.polynomial.Polynomial.fromroots(np.delete(x, i)) / (x[i] - np.delete(x, i)).prod()
    
    return poly


def newton(x, y):
    def iter(knot):
        res = np.longdouble(0)
        for i in range(knot + 1):
            res += y[i] / (x[i] - np.delete(x[:knot + 1], i)).prod()
        return res
    
    poly = y[0] * np.polynomial.Polynomial([1])
    for i in range(1, x.shape[0]):
        poly += iter(i) * np.polynomial.Polynomial.fromroots(x[:i])
    
    return poly
```

# Polynomial

```python
def polynomial(x):
    return np.exp(-4 * np.sin(x))
```

# Plot

```python
def draw_plot(a, b, n, f, dist, inter, draw):
    n1 = 500
    x1 = np.linspace(a, b, n1)
    y1 = f(x1)
    x = np.linspace(a, b, n1)
    xf = dist(a, b, n)
    yp = [f(xf[i]) for i in range(len(xf))]
    yf = inter(xf, yp)(x)
    
    if draw:
        plt.scatter(xf, yp, color="black")
        plt.plot(x, yf,color="darkmagenta")  # given polynomial
        plt.plot(x1, y1, color='mediumvioletred') # my function
        plt.show()
    
    error = [abs(y1[i] - yf[i]) for i in range(n1)]
    max_e = 0
    sq_error = 0
    for i in range(n1): 
        max_e = max(max_e, error[i])
        sq_error += error[i]**2
    sq_error /= n1
    print("max_err, \t\tsq_err, \t\tknots")
    print(f"{max_e}, \t{sq_error}, \t{n}")
    return max_e, sq_error, n
    
```

# Main function

```python
start = -4 * np.pi
end = 3 * np.pi

knots_numbers = [3, 4, 5, 7, 10, 15, 20, 30, 40, 50]

for knots_nb in knots_numbers:
    print(f"\n\n\nIlość węzłów: {knots_nb}, rozmieszczenie równomierne, metoda Lagrange'a\n")
    draw_plot(start, end, knots_nb, polynomial, even_distribution, lagrange, True)
    
    print(f"\n\n\nIlość węzłów: {knots_nb}, rozmieszczenie równomierne, metoda Newtona\n")
    draw_plot(start, end, knots_nb, polynomial, even_distribution, newton, True)
    
    print(f"\n\n\nIlość węzłów: {knots_nb}, rozmieszczenie Czebysheva, metoda Lagrange'a\n")
    draw_plot(start, end, knots_nb, polynomial, chebyshev_distribution, lagrange, True)
    
    print(f"\n\n\nIlość węzłów: {knots_nb}, rozmieszczenie Czebysheva, metoda Newtona\n")
    draw_plot(start, end, knots_nb, polynomial, chebyshev_distribution, newton, True)
    
```
